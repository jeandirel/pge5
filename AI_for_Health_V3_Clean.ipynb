| Dimension | Standard AI | Healthcare AI |
|-----------|-------------|---------------|

# DNA Feature Extraction

---

## 3.1 Background — DNA as Data

DNA (Deoxyribonucleic Acid) encodes the genetic instructions of every living organism. From a computational perspective, it is a sequence over a four-symbol alphabet:

| Symbol | Nucleotide | Complementary Base | Hydrogen Bonds |
|--------|-----------|-------------------|----------------|
| A | Adenine | T | 2 — weaker pair |
| T | Thymine | A | 2 — weaker pair |
| G | Guanine | C | 3 — stronger pair |
| C | Cytosine | G | 3 — stronger pair |

**The core machine learning problem.** A model cannot accept the string `"ATGCGATCG"` as input. It requires a fixed-length numerical vector. The process of transforming a sequence into such a vector is called *feature engineering*, and it must preserve as much biologically meaningful information as possible.

---

## 3.2 Feature Definitions

<div style="font-family:'Segoe UI',Arial,sans-serif; background:#f4f6f8; border:1px solid #cfd8dc;
            border-radius:6px; padding:20px 28px; margin:14px 0;">

| Feature | Formula | Biological Meaning |
|---------|---------|-------------------|
| GC Content | (G + C) / total | Thermal stability — G-C pairs have 3 H-bonds |
| AT Content | (A + T) / total | Complement of GC content |
| Nucleotide frequencies | count(X) / total | Compositional bias |
| k-mer frequencies | count(substring length k) / total | Local sequence patterns |
| CpG ratio | observed CpG / expected CpG | Epigenetic methylation marker |
| Sequence length | len(sequence) | Gene size |

</div>

**GC Content formula:**

$$GC\ (\%) = \frac{G + C}{A + T + G + C} \times 100$$

<div class="nb-insight">
<strong>Biological note.</strong> G-C base pairs form three hydrogen bonds compared to two for A-T pairs. Organisms adapted to high-temperature environments (thermophilic bacteria) tend to have higher GC content because the additional bond increases DNA thermal stability. This makes GC content a meaningful, interpretable feature for species classification and gene prediction.
</div>

---

## 3.3 Implementation

**Step 1 — Basic sequence analysis**


```python
# ── Step 1: Basic nucleotide analysis ──────────────────────────────────────

sequence = "ATGCGATCGGCAATCGTAGCGATCGAATCGATCGATCGATCGATCGCGATCGGCAT"

total = len(sequence)
A = sequence.count('A')
T = sequence.count('T')
G = sequence.count('G')
C = sequence.count('C')

gc_content = (G + C) / total
at_content = (A + T) / total

print("Sequence Analysis")
print("─" * 46)
print(f"  Sequence   : {sequence}")
print(f"  Length     : {total} bp")
print()
print(f"  {'Base':<12} {'Count':>6}  {'Frequency':>10}")
print(f"  {'────':<12} {'─────':>6}  {'─────────':>10}")
for base, count in zip('ATGC', [A, T, G, C]):
    print(f"  {base:<12} {count:>6}  {count/total:>10.1%}")
print()
print(f"  GC content : {gc_content:.4f}  ({gc_content*100:.1f}%)")
print(f"  AT content : {at_content:.4f}  ({at_content*100:.1f}%)")
print(f"  Verification: GC + AT = {gc_content + at_content:.4f}")
```

**Step 2 — Fundamental biological operations**

```python
# ── Step 2: Complement, reverse complement, transcription ──────────────────

complement_table = str.maketrans('ATGC', 'TACG')

complement         = sequence.translate(complement_table)
reverse_complement = complement[::-1]
mrna               = sequence.replace('T', 'U')   # DNA -> mRNA

print("Biological Sequence Operations")
print("─" * 60)
print(f"  Original DNA (5'->3')      : {sequence[:45]}...")
print(f"  Complement   (3'->5')      : {complement[:45]}...")
print(f"  Reverse complement (5'->3'): {reverse_complement[:45]}...")
print(f"  mRNA transcript            : {mrna[:45]}...")
print()
print("Note: The reverse complement is the second DNA strand,")
print("written in the conventional 5'->3' direction.")
```

**Step 3 — k-mer extraction and frequency analysis**

A *k-mer* is a substring of length *k* drawn from a biological sequence. For k=3 and sequence `ATGCAT`, the 3-mers are `ATG`, `TGC`, `GCA`, `CAT`. k-mers capture **local sequence context** that simple composition counts miss.

Applications: genome assembly, pathogen identification, antibiotic resistance detection, cancer mutation signature analysis.


```python
# ── Step 3: k-mer extraction ───────────────────────────────────────────────

def extract_kmers(seq, k):
    return [seq[i:i+k] for i in range(len(seq) - k + 1)]

def kmer_frequencies(seq, k):
    kmers = extract_kmers(seq, k)
    total = len(kmers)
    return {km: cnt/total for km, cnt in Counter(kmers).most_common()}

freq_2 = kmer_frequencies(sequence, k=2)   # dinucleotides
freq_3 = kmer_frequencies(sequence, k=3)   # codons

print("Dinucleotide frequencies (k=2):")
print("─" * 36)
for km, f in list(freq_2.items())[:8]:
    bar = chr(9608) * int(f * 60)
    print(f"  {km}  {f:.3f}  {bar}")

print()
print("Trinucleotide (codon) frequencies (k=3) — top 8:")
print("─" * 36)
for km, f in list(freq_3.items())[:8]:
    bar = chr(9608) * int(f * 80)
    print(f"  {km}  {f:.3f}  {bar}")

print()
print("Note: ATG is the universal start codon.")
print("Its over-representation may indicate coding sequence content.")
```

**Step 4 — Build the ML-ready feature table**

```python
# ── Step 4: Multi-sequence feature table ───────────────────────────────────

def compute_dna_features(seq_id, seq):
    seq = seq.upper().replace(' ', '')
    n   = len(seq)
    if n == 0:
        return None
    A = seq.count('A'); T = seq.count('T')
    G = seq.count('G'); C = seq.count('C')
    gc  = (G + C) / n
    km2 = kmer_frequencies(seq, 2)
    cpg_obs = seq.count('CG')
    cpg_exp = (C * G) / n if n > 0 else 0
    cpg_ratio = cpg_obs / cpg_exp if cpg_exp > 0 else 0
    return {
        'seq_id'     : seq_id,
        'length'     : n,
        'freq_A'     : round(A/n, 4),
        'freq_T'     : round(T/n, 4),
        'freq_G'     : round(G/n, 4),
        'freq_C'     : round(C/n, 4),
        'gc_content' : round(gc, 4),
        'at_content' : round(1 - gc, 4),
        'cpg_ratio'  : round(cpg_ratio, 4),
        'dinuc_CG'   : round(km2.get('CG', 0), 4),
        'dinuc_AT'   : round(km2.get('AT', 0), 4),
    }

sequences = {
    'GENE_001' : 'ATGCGATCGGCAATCGTAGCGATCGAATCGATCGATCGATCGATCGCGATCGGCAT',
    'GENE_002' : 'GCGCGCGCGCGCGCATATATATATATATAGCGCGCGCGCGCGCGCGCGCGCGCGCG',
    'GENE_003' : 'ATATATATATATATATATATATATATATATATATATATATAT',
    'GENE_004' : 'ATGAAAGCATTTTTTTTAAAAAAAAATTTTTTAGCTAGCTAGCTAGCTAGCTAGCT',
    'GENE_005' : 'CGCGCGCGATCGATCGATCGCGCGATCGATCGATCGATCGATCGCGCGATCGATCG',
    'GENE_006' : 'TTAGGGTTAGGGTTAGGGTTAGGGTTAGGGTTAGGG',
    'GENE_007' : 'ATGATGATGATGATGATGATGATGATGATGATGATGATGATGATGATGATGATG',
}

rows = [compute_dna_features(k, v) for k, v in sequences.items()]
df_dna = pd.DataFrame(rows).set_index('seq_id')

print("DNA Feature Table")
print("─" * 72)
print(df_dna.to_string())
print()
print(f"Shape: {df_dna.shape[0]} sequences x {df_dna.shape[1]} features")
print("This DataFrame is now directly usable as input to any ML model.")
```

```python
# ── Step 5: Visualisation ──────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("TP1 — DNA Feature Analysis", fontsize=13, fontweight='bold',
             color=BLUE, y=1.02)

# GC Content bar chart
gc_vals    = df_dna['gc_content']
bar_colors = [TEAL if v >= 0.5 else BMID for v in gc_vals]
bars = axes[0].bar(df_dna.index, gc_vals, color=bar_colors,
                    edgecolor='white', linewidth=1.2, width=0.6)
axes[0].axhline(0.5, color=AMBER, linestyle='--', linewidth=1.2, label='50% threshold')
axes[0].set_title("GC Content by Sequence", fontweight='bold')
axes[0].set_xlabel("Sequence identifier")
axes[0].set_ylabel("GC content (fraction)")
axes[0].set_ylim(0, 1.05)
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(framealpha=0.9)
for bar, val in zip(bars, gc_vals):
    axes[0].text(bar.get_x() + bar.get_width()/2, val + 0.02,
                  f"{val:.2f}", ha='center', fontsize=8.5,
                  fontweight='bold', color=BLUE)

# Nucleotide frequency heatmap
freq_df = df_dna[['freq_A','freq_T','freq_G','freq_C']].copy()
freq_df.columns = ['A', 'T', 'G', 'C']
sns.heatmap(freq_df, ax=axes[1], cmap='Blues', annot=True, fmt='.2f',
            linewidths=0.8, linecolor='white', vmin=0, vmax=0.8,
            annot_kws={'size': 9, 'weight': 'bold', 'color': BLUE},
            cbar_kws={'label': 'Frequency', 'shrink': 0.8})
axes[1].set_title("Nucleotide Frequency Heatmap", fontweight='bold')
axes[1].set_xlabel("Nucleotide")
axes[1].set_ylabel("")

plt.tight_layout()
plt.savefig('tp1_dna_analysis.png', dpi=120, bbox_inches='tight')
plt.show()
```

# Machine Learning Pipeline for Health Prediction

---

## 4.1 Supervised Learning — Core Concepts

In supervised learning, a model learns a mapping from input features to output labels by training on labelled examples.

| Concept | Definition | Clinical Example |
|---------|------------|-----------------|
| **Feature (X)** | A measurable input variable | Systolic blood pressure: 145 mmHg |
| **Label (y)** | The quantity to predict | 1 = cardiovascular event within 5 years |
| **Training set** | Data used to fit the model | 80% of patient records |
| **Test set** | Data used to evaluate the model | 20% never seen during training |
| **Overfitting** | Memorises training data, fails on new cases | Train F1 = 0.98, Test F1 = 0.54 |
| **Generalisation** | Performs well on genuinely new data | Train F1 = 0.87, Test F1 = 0.84 |

**Pipeline overview:**

<div style="font-family:'Segoe UI',Arial,sans-serif; background:#f4f6f8; border:1px solid #cfd8dc;
            border-radius:6px; padding:18px 24px; margin:14px 0; max-width:600px;">

| Step | Action |
|------|--------|
| 1 | Define features and target |
| 2 | Split into train / test (stratified) |
| 3 | Scale features (StandardScaler) |
| 4 | Train model on training set only |
| 5 | Predict on test set |
| 6 | Compute clinical evaluation metrics |

</div>

<div class="nb-critical">
<strong>Data leakage warning.</strong> Never use the test set for any decision made during training — this includes feature selection, threshold tuning, and scaling parameter estimation. The scaler must be fitted on training data only and then applied to the test set. Leakage inflates reported performance and has contributed to several high-profile AI failures in clinical settings.
</div>

## 4.2 The Synthetic Clinical Dataset

## 4.3 Why Accuracy is Insufficient in Healthcare

<div class="nb-critical">
<strong>The class imbalance trap.</strong> Consider a cancer screening dataset where 5% of patients are positive. A classifier that always predicts "negative" achieves 95% accuracy while detecting zero cancers. Its recall is 0%. In clinical practice, this model is useless and dangerous — yet a naive accuracy-based evaluation would label it nearly perfect.
</div>

| Metric | Formula | Clinical Priority |
|--------|---------|-------------------|
| **Recall (Sensitivity)** | TP / (TP + FN) | Highest — do not miss sick patients |
| **Precision (PPV)** | TP / (TP + FP) | High — avoid unnecessary intervention |
| **F1-Score** | 2 × Precision × Recall / (P + R) | Use when both matter equally |
| **Specificity** | TN / (TN + FP) | Important in low-prevalence screening |
| **Accuracy** | (TP + TN) / total | Only valid when classes are balanced |

# ECG Signal Analysis

## 5.1 Background — The Electrocardiogram

An electrocardiogram (ECG) records the electrical activity of the heart through electrodes placed on the skin. Each cardiac cycle produces a characteristic waveform composed of identifiable components.

| Wave / Interval | Origin | Clinical Significance |
|----------------|--------|----------------------|
| P wave | Atrial depolarisation | Atria contracting; abnormal if absent or irregular |
| PR interval | AV node conduction delay | Prolonged in first-degree heart block |
| QRS complex | Ventricular depolarisation | Widened in bundle branch block; morphology changes in MI |
| ST segment | Early ventricular repolarisation | Elevation indicates myocardial infarction |
| T wave | Ventricular repolarisation | Inversion suggests ischaemia or electrolyte disturbance |
| QT interval | Full ventricular cycle | Prolongation — Long QT syndrome, drug toxicity |
| RR interval | Time between successive R-peaks | Basis of heart rate and HRV calculation |

**Processing pipeline:**

<div style="font-family:'Segoe UI',Arial,sans-serif; background:#f4f6f8; border:1px solid #cfd8dc;
            border-radius:6px; padding:18px 24px; margin:14px 0; max-width:580px;">

| Step | Operation |
|------|-----------|
| 1 | Acquire raw signal (360 Hz) |
| 2 | Bandpass filter (0.5 – 40 Hz) — remove baseline wander and high-frequency noise |
| 3 | R-peak detection — Pan-Tompkins or derivative-based algorithm |
| 4 | Extract RR intervals |
| 5 | Compute heart rate and HRV indices (SDNN, RMSSD) |
| 6 | Extract morphological features (QRS width, ST level) |
| 7 | Feed feature vector into classifier |

</div>

<div class="nb-info">
<strong>MIT-BIH Arrhythmia Database.</strong> Published on PhysioNet, this database contains 48 half-hour two-lead ECG recordings at 360 Hz with expert cardiologist annotations. It is the gold-standard benchmark for ECG algorithm development and validation. The <code>wfdb</code> Python library enables direct programmatic access.
</div>

```python
# ── TP3 Step 1: Load ECG data ───────────────────────────────────────────────
try:
    import wfdb
    record = wfdb.rdrecord('100', sampfrom=0, sampto=3600, pb_dir='mitdb')
    annot  = wfdb.rdann('100', 'atr', sampfrom=0, sampto=3600, pb_dir='mitdb')
    ecg_signal = record.p_signal[:, 0]
    FS         = record.fs
    print(f"MIT-BIH record 100 loaded: {FS} Hz, {len(ecg_signal)} samples")
    print(f"Expert annotations: {len(annot.sample)} beats, types: {set(annot.symbol)}")
    USE_REAL = True
except Exception as e:
    print(f"wfdb unavailable ({type(e).__name__}). Generating synthetic ECG.")
    USE_REAL = False

# ── Synthetic ECG (always generated for reproducible demonstration) ──────────
def synthetic_ecg(duration=10, fs=360, bpm=72, noise_std=0.03):
    t  = np.linspace(0, duration, int(duration * fs))
    sg = np.zeros_like(t)
    bi = int(60 / bpm * fs)
    peaks = []
    for b0 in range(0, len(t) - bi, bi):
        pc = b0 + int(0.27 * fs)
        tc = b0 + int(0.45 * fs)
        pw = int(0.04 * fs)
        tw = int(0.07 * fs)
        for i in range(max(0, b0 + int(0.12*fs) - 3*pw),
                        min(len(t), b0 + int(0.12*fs) + 3*pw)):
            sg[i] += 0.15 * np.exp(-0.5 * ((i - (b0+int(0.12*fs))) / pw)**2)
        for i in range(max(0, tc - 3*tw), min(len(t), tc + 3*tw)):
            sg[i] += 0.30 * np.exp(-0.5 * ((i - tc) / tw)**2)
        if pc < len(sg):
            q = pc - int(0.02 * fs); s = pc + int(0.02 * fs)
            if 0 <= q < len(sg): sg[q] -= 0.10
            sg[pc] += 1.0; peaks.append(pc)
            if 0 <= s < len(sg): sg[s] -= 0.25
    sg += np.random.normal(0, noise_std, len(t))
    sg += 0.05 * np.sin(2 * np.pi * 0.15 * t)   # baseline wander
    return t, sg, np.array(peaks)

t_ecg, ecg_raw, r_peaks = synthetic_ecg(duration=10, fs=360, bpm=72)
FS_DEMO = 360

print(f"Synthetic ECG ready: {len(ecg_raw)} samples at {FS_DEMO} Hz, {len(r_peaks)} beats")
```

# EEG Signal Analysis and Classification

---

## 6.1 Background — The Electroencephalogram

An electroencephalogram (EEG) records the summed electrical activity of cortical neurons through electrodes on the scalp. The signal is small (10–100 microvolts), non-stationary, and subject to numerous artefacts.

**Frequency bands and their interpretation:**

| Band | Range | Brain State | Clinical Application |
|------|-------|-------------|---------------------|
| Delta (delta) | 0.5 – 4 Hz | Deep sleep, anaesthesia | Sleep staging, coma monitoring |
| Theta (theta) | 4 – 8 Hz | Drowsiness, working memory | ADHD assessment, cognitive load |
| Alpha (alpha) | 8 – 13 Hz | Relaxed, eyes closed | Neurofeedback, stress monitoring |
| Beta (beta) | 13 – 30 Hz | Active cognition, motor control | Anxiety, BCI motor imagery |
| Gamma (gamma) | 30 – 100 Hz | High-level perception | Consciousness research |

**Principal artefact sources:**

| Artefact | Frequency Content | Mitigation |
|----------|-------------------|-----------|
| Ocular (blinks) | < 4 Hz, large amplitude | ICA, regression |
| Muscular (EMG) | > 30 Hz, broadband | Notch filter, ICA |
| Power line | 50 Hz or 60 Hz, narrowband | Notch filter |
| Cardiac (ECG) | ~1 Hz + harmonics | ICA |
| Electrode movement | Broadband spikes | Rejection by threshold |

**Feature extraction method — Welch Power Spectral Density:**

<div style="font-family:'Segoe UI',Arial,sans-serif; background:#f4f6f8; border:1px solid #cfd8dc;
            border-radius:6px; padding:16px 22px; margin:12px 0; max-width:560px;">

| Step | Operation |
|------|-----------|
| 1 | Segment signal into 2-second windows (50% overlap) |
| 2 | Apply Welch PSD to each window per channel |
| 3 | Integrate PSD within each frequency band |
| 4 | Assemble feature vector: 14 channels x 5 bands = 70 features |
| 5 | Train Random Forest classifier |

</div>

<div class="nb-insight">
<strong>Why Welch PSD?</strong> Simple FFT applied to short, non-stationary signals produces noisy spectral estimates. Welch's method averages multiple overlapping windowed FFTs, producing a smoother, more stable power estimate. This is standard practice in clinical EEG analysis.
</div>

---


```python
# ── TP4 Step 1: Generate synthetic EEG dataset ──────────────────────────────
from scipy.signal import welch
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, classification_report)

def generate_eeg_dataset(n_samples=2000, fs=128, n_channels=14, seed=42):
    '''
    Synthetic EEG: binary label — 0=eyes open, 1=eyes closed.
    Eyes closed: dominant alpha (10 Hz, higher amplitude).
    Eyes open  : dominant beta  (20 Hz, higher amplitude).
    '''
    np.random.seed(seed)
    t   = np.arange(n_samples) / fs
    lbl = np.random.randint(0, 2, n_samples)
    channels = {}
    for ch in range(n_channels):
        s = np.zeros(n_samples)
        for i in range(n_samples):
            alpha_amp = np.random.uniform(4, 9)   if lbl[i] == 1 else np.random.uniform(0.5, 2)
            beta_amp  = np.random.uniform(0.5, 2) if lbl[i] == 1 else np.random.uniform(3, 7)
            s[i] = (alpha_amp * np.sin(2*np.pi*10*t[i] + np.random.randn()) +
                    beta_amp  * np.sin(2*np.pi*20*t[i] + np.random.randn()) +
                    np.random.normal(0, 0.4))
        channels[f'CH{ch+1:02d}'] = s
    df = pd.DataFrame(channels)
    df['label'] = lbl
    return df

df_eeg = generate_eeg_dataset()
print(f"EEG dataset: {len(df_eeg)} samples, 14 channels, 128 Hz")
print(f"  Eyes open  (0): {(df_eeg['label']==0).sum()}")
print(f"  Eyes closed(1): {(df_eeg['label']==1).sum()}")
```

```python
# ── TP4 Step 2: Band power feature extraction ────────────────────────────────

BANDS    = {'delta':(0.5,4), 'theta':(4,8), 'alpha':(8,13), 'beta':(13,30), 'gamma':(30,45)}
FS_EEG   = 128
CHANNELS = [f'CH{ch+1:02d}' for ch in range(14)]
WIN_SIZE = 256   # 2 seconds
STEP     = 128   # 50% overlap

def compute_band_power(signal_1d, fs, band):
    freqs, psd = welch(signal_1d, fs=fs, nperseg=min(256, len(signal_1d)))
    mask = (freqs >= band[0]) & (freqs <= band[1])
    return float(np.trapz(psd[mask], freqs[mask]))

print("Extracting band power features...")
feature_rows, label_rows = [], []

for start in range(0, len(df_eeg) - WIN_SIZE, STEP):
    window = df_eeg.iloc[start:start + WIN_SIZE]
    label  = int(window['label'].mode()[0])
    row    = {}
    for ch in CHANNELS:
        for band_name, band_range in BANDS.items():
            row[f'{ch}_{band_name}'] = compute_band_power(
                window[ch].values, FS_EEG, band_range)
    feature_rows.append(row)
    label_rows.append(label)

df_features = pd.DataFrame(feature_rows)
y_eeg       = np.array(label_rows)

print(f"Feature matrix: {df_features.shape[0]} windows x {df_features.shape[1]} features")
print(f"(14 channels x 5 bands = {14*5} features per window)")
```

```python
# ── TP4 Step 3: Classification and evaluation ────────────────────────────────

X_eeg = df_features.values
X_tr, X_te, y_tr, y_te = train_test_split(
    X_eeg, y_eeg, test_size=0.2, random_state=42, stratify=y_eeg)

rf_model = RandomForestClassifier(
    n_estimators=150, max_depth=7,
    class_weight='balanced', random_state=42, n_jobs=-1)
rf_model.fit(X_tr, y_tr)
y_pr = rf_model.predict(X_te)

print("EEG Classification — Test Set Results")
print("─" * 46)
print(f"  Accuracy  : {accuracy_score(y_te, y_pr)*100:.1f}%")
print(f"  Precision : {precision_score(y_te, y_pr)*100:.1f}%")
print(f"  Recall    : {recall_score(y_te, y_pr)*100:.1f}%")
print(f"  F1-Score  : {f1_score(y_te, y_pr)*100:.1f}%")
print()
print(classification_report(y_te, y_pr, target_names=['Eyes Open', 'Eyes Closed']))
```

```python
# ── TP4 Step 4: Feature importance visualisation ─────────────────────────────

importances = pd.Series(rf_model.feature_importances_, index=df_features.columns)
top_20      = importances.nlargest(20)

band_color_map = {'alpha': BMID, 'beta': TEAL, 'theta': GRAY,
                   'delta': RED,  'gamma': AMBER}

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle("TP4 — EEG Feature Importance", fontsize=13, fontweight='bold', color=BLUE)

# Top-20 horizontal bar
feat_colors = [band_color_map.get(name.split('_')[1], GRAY) for name in top_20.index]
axes[0].barh(range(len(top_20)), top_20.values,
              color=feat_colors, edgecolor='white', linewidth=0.8)
axes[0].set_yticks(range(len(top_20)))
axes[0].set_yticklabels(top_20.index, fontsize=8)
axes[0].invert_yaxis()
axes[0].set_title("Top 20 Features by Importance", fontweight='bold')
axes[0].set_xlabel("Mean decrease in impurity")

# Legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=v, label=k.capitalize())
                    for k, v in band_color_map.items()]
axes[0].legend(handles=legend_elements, fontsize=8, loc='lower right')

# Band-level importance pie
band_importance = {bn: importances[importances.index.str.contains(bn)].sum()
                    for bn in BANDS}
pie_colors = [band_color_map[b] for b in band_importance]
wedges, texts, autotexts = axes[1].pie(
    band_importance.values(),
    labels=[f"{k.capitalize()}\n({v:.3f})" for k, v in band_importance.items()],
    colors=pie_colors, autopct='%1.1f%%',
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    textprops={'fontsize': 9, 'color': BLUE})
for at in autotexts: at.set_fontweight('bold')
axes[1].set_title("Band Power Contribution", fontweight='bold')

plt.tight_layout()
plt.savefig('tp4_eeg.png', dpi=120, bbox_inches='tight')
plt.show()

print()
print("Alpha band dominates — consistent with the alpha-blocking phenomenon:")
print("alpha power reliably decreases when the eyes are opened.")
```

# Ethics in Healthcare AI

---

## 7.1 Why Ethical Requirements are Elevated in Healthcare

<div class="nb-critical">
When an AI system makes an erroneous recommendation for a film, the user is mildly inconvenienced. When an AI system makes an erroneous clinical prediction, a patient may receive wrong treatment, delayed diagnosis, or unnecessary intervention — consequences that can be severe or irreversible. The engineer who designs such a system bears partial moral and, increasingly, legal responsibility for its downstream effects.
</div>

---

## 7.2 Seven Pillars of Responsible Healthcare AI

| Pillar | Definition | Practical Implication |
|--------|------------|----------------------|
| **Privacy** | Patient data is legally protected (GDPR, HIPAA) | Federated learning, differential privacy, de-identification |
| **Fairness** | Equal performance across demographic groups | Audit sub-group metrics; oversample under-represented populations |
| **Transparency** | Stakeholders can understand how decisions are made | Document data sources, model type, known limitations |
| **Explainability** | Clinicians understand why a prediction was made | SHAP values, LIME, decision trees, attention maps |
| **Accountability** | Defined responsibility when the AI causes harm | Mandatory human-in-the-loop; never fully autonomous diagnosis |
| **Safety** | Clinically validated before patient exposure | Prospective trials; FDA 510(k) or EU CE Mark |
| **Equity** | AI must not amplify existing health disparities | Monitor performance across all demographic strata |

---

## 7.3 Documented AI Failures in Healthcare

| Case | Root Cause | Consequence |
|------|-----------|------------|
| Skin cancer classifier (2019) | Training data lacked diversity — mostly light skin | Near-random performance on dark skin tones |
| Commercial sepsis algorithm | Race-adjusted eGFR formula embedded systemic bias | Systematic underestimation of risk in Black patients |
| Chest X-ray AI (multiple) | Over-representation of middle-aged male training subjects | Significantly reduced sensitivity in women and elderly patients |
| Hospital-specific sepsis model | Learned local documentation patterns | Performance collapsed when deployed at other institutions |

<div class="nb-warning">
<strong>Dataset is society.</strong> If training labels were assigned by a clinician whose practice embedded historical biases, the model will learn and amplify those biases — not medical truth. Auditing the label generation process is as important as auditing the features.
</div>

```python
# ── Ethics Cell: Permutation feature importance ─────────────────────────────
# Permutation importance quantifies each feature's contribution by measuring
# how much model performance degrades when that feature is randomly shuffled.
# This is a model-agnostic explainability method accessible to clinicians.

from sklearn.inspection import permutation_importance

perm = permutation_importance(
    model, X_test_s, y_test,
    n_repeats=25, random_state=42, scoring='f1')

imp_df = pd.DataFrame({
    'feature'    : FEATURES,
    'importance' : perm.importances_mean,
    'std'        : perm.importances_std
}).sort_values('importance', ascending=False)

print("Feature Importance — Permutation Method (F1 metric)")
print("─" * 52)
print("Interpretation: larger value = greater impact on prediction quality")
print()
print(imp_df.to_string(index=False))

# Visualisation
fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle("Ethics — Model Explainability: Permutation Feature Importance",
             fontsize=12, fontweight='bold', color=BLUE)

bar_colors = [RED if v > 0.05 else BMID if v > 0 else LGRAY
               for v in imp_df['importance']]
ax.barh(imp_df['feature'], imp_df['importance'],
         xerr=imp_df['std'], color=bar_colors,
         edgecolor='white', capsize=4, height=0.55)
ax.axvline(0, color=GRAY, ls='--', lw=0.9)
ax.set_xlabel("Mean decrease in F1-score when feature is permuted")
ax.set_title("Features with positive importance are genuinely predictive",
              fontsize=10, color=GRAY)
ax.invert_yaxis()

from matplotlib.patches import Patch
ax.legend(handles=[
    Patch(color=RED,   label='High importance  (> 0.05)'),
    Patch(color=BMID,  label='Moderate importance'),
    Patch(color=LGRAY, label='Low / negligible importance'),
], fontsize=9)

plt.tight_layout()
plt.savefig('ethics_explainability.png', dpi=120, bbox_inches='tight')
plt.show()
print()
print("This chart can be shown to clinicians to validate model behaviour.")
print("EU AI Act (Article 13) requires that high-risk AI systems provide")
print("sufficient transparency for human oversight.")
```

## 8.4 Consolidated Key Principles

**Bioinformatics**
- DNA requires numerical encoding; raw character sequences cannot enter a machine learning model directly.
- GC content and k-mer frequencies are biologically interpretable features with well-established meaning.

**Machine Learning for Health**
- The train/test split must be performed before any preprocessing parameter is estimated.
- Recall is the primary evaluation metric in most diagnostic and screening tasks.
- A model's accuracy is only meaningful when the class distribution is balanced.

**ECG Analysis**
- R-peak detection is the foundation of all time-domain ECG feature extraction.
- HRV indices (SDNN, RMSSD) encode autonomic nervous system activity and are clinically validated biomarkers.

**EEG Analysis**
- Frequency-band power is the most robust and interpretable EEG feature.
- Alpha suppression upon eye-opening is a well-established neurophysiological marker.
- Cross-subject generalisation remains the central open problem in EEG-based AI.

**Ethics**
- Recall over accuracy is not just a technical preference — it is an ethical requirement.
- Explainability is a clinical safety requirement, not an optional enhancement.
- The engineer is a participant in the healthcare system and carries corresponding responsibility.
